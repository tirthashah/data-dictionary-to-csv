```python
from PIL import Image

img = Image.open('untitled.jpg')
img
```

```python
#convert image into text
text = pytesseract.image_to_string(img,lang='eng')
text

```

```python
#clean the text
cleaned_text = text.replace('\n', ',')
clean=cleaned_text.replace(',,',',')
c=clean.replace('Star1,','')
print(c)
```

```python
# Create an empty dictionary to store the text
text_dict = {}

# Split the text into lines and add them to the dictionary
for i, line in enumerate(c.split(',Star2,')):
    text_dict[i] = line


# Print the resulting dictionary
print(text_dict)
```
